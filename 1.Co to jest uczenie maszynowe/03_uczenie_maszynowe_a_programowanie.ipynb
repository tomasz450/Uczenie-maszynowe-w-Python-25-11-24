### 1.1. Porwnanie ML z tradycyjnym programowaniem

1. Podstawowa rnica w podejciu:

   - Tradycyjne programowanie: Programista definiuje reguy, ktre przetwarzaj dane na wyniki.
   - Uczenie maszynowe: System uczy si regu na podstawie danych i podanych wynikw.
   
   
2. Schemat dziaania:

   - Tradycyjne: Dane + Reguy (algorytm) = Wyniki
   - Uczenie maszynowe: Dane + Wyniki (etykiety) = Reguy (model) -> Dane + Model = Wyniki


3. Rola programisty:

   - Tradycyjne: Programista dokadnie okrela kady krok algorytmu.
   - Uczenie maszynowe: Programista przygotowuje dane, wybiera i konfiguruje model, ale nie definiuje bezporednio regu decyzyjnych.


4. Zdolno do adaptacji:

   - Tradycyjne: Ograniczona do scenariuszy przewidzianych przez programist.
   - Uczenie maszynowe: Moe adaptowa si do nowych, niewidzianych wczeniej danych*.


5. Obsuga zoonoci:
   - Tradycyjne: Trudnoci w obsudze bardzo zoonych problemw z wieloma zmiennymi.
   - Uczenie maszynowe: Lepiej radzi sobie ze zoonymi wzorcami i nieliniowymi zalenociami.


6. Przykady zada:
   - Tradycyjne: Sortowanie listy, obliczanie podatku na podstawie jasno zdefiniowanych regu.
   - Uczenie maszynowe: Rozpoznawanie obrazw, prognozowanie cen, wykrywanie anomalii.


7. Podejcie do rozwizywania problemw:
   - Tradycyjne: Dekompozycja problemu na mniejsze, dokadnie zdefiniowane kroki.
   - Uczenie maszynowe: Znalezienie odpowiednich danych i pozwolenie modelowi na odkrycie wzorcw.


8. Przewidywalno wynikw:
    - Tradycyjne: Deterministyczne, zawsze te same wyniki dla tych samych danych wejciowych.
    - Uczenie maszynowe: Moe wprowadza element losowoci, wyniki mog si nieznacznie rni.
    
    *tutaj take jest dostrzegalna rnica w projektach IT vs ML (wycena zada, oszacowanie wynikw - trudniejsze)

9. Interpretacja dziaania:
    - Tradycyjne: atwe do zrozumienia i wytumaczenia (white-box).
    - Uczenie maszynowe: Czsto trudne do interpretacji, szczeglnie w przypadku zoonych modeli (black-box).


10. Wymagania sprztowe:
    - Tradycyjne: Zazwyczaj mniejsze wymagania obliczeniowe.
    - Uczenie maszynowe: Czsto wymaga znacznych zasobw obliczeniowych, szczeglnie w fazie treningu.

### 1.1. Porównanie ML z tradycyjnym programowaniem

1. Podstawowa różnica w podejściu:

   - Tradycyjne programowanie: Programista definiuje reguły, które przetwarzają dane na wyniki.
   - Uczenie maszynowe: System uczy się reguł na podstawie danych i pożądanych wyników.
   
   
2. Schemat działania:

   - Tradycyjne: Dane + Reguły (algorytm) = Wyniki
   - Uczenie maszynowe: Dane + Wyniki (etykiety) = Reguły (model) -> Dane + Model = Wyniki


3. Rola programisty:

   - Tradycyjne: Programista dokładnie określa każdy krok algorytmu.
   - Uczenie maszynowe: Programista przygotowuje dane, wybiera i konfiguruje model, ale nie definiuje bezpośrednio reguł decyzyjnych.


4. Zdolność do adaptacji:

   - Tradycyjne: Ograniczona do scenariuszy przewidzianych przez programistę.
   - Uczenie maszynowe: Może adaptować się do nowych, niewidzianych wcześniej danych*.


5. Obsługa złożoności:
   - Tradycyjne: Trudności w obsłudze bardzo złożonych problemów z wieloma zmiennymi.
   - Uczenie maszynowe: Lepiej radzi sobie ze złożonymi wzorcami i nieliniowymi zależnościami.


6. Przykłady zadań:
   - Tradycyjne: Sortowanie listy, obliczanie podatku na podstawie jasno zdefiniowanych reguł.
   - Uczenie maszynowe: Rozpoznawanie obrazów, prognozowanie cen, wykrywanie anomalii.


7. Podejście do rozwiązywania problemów:
   - Tradycyjne: Dekompozycja problemu na mniejsze, dokładnie zdefiniowane kroki.
   - Uczenie maszynowe: Znalezienie odpowiednich danych i pozwolenie modelowi na odkrycie wzorców.


8. Przewidywalność wyników:
    - Tradycyjne: Deterministyczne, zawsze te same wyniki dla tych samych danych wejściowych.
    - Uczenie maszynowe: Może wprowadzać element losowości, wyniki mogą się nieznacznie różnić.
    
    *tutaj także jest dostrzegalna różnica w projektach IT vs ML (wycena zadań, oszacowanie wyników - trudniejsze)

9. Interpretacja działania:
    - Tradycyjne: Łatwe do zrozumienia i wytłumaczenia (white-box).
    - Uczenie maszynowe: Często trudne do interpretacji, szczególnie w przypadku złożonych modeli (black-box).


10. Wymagania sprzętowe:
    - Tradycyjne: Zazwyczaj mniejsze wymagania obliczeniowe.
    - Uczenie maszynowe: Często wymaga znacznych zasobów obliczeniowych, szczególnie w fazie treningu.

### Hiperparametry modelu:

1. Definicja:
   - Hiperparametry to parametry, ktre s ustawiane przed rozpoczciem procesu uczenia modelu.
   - W przeciwiestwie do parametrw modelu, hiperparametry nie s uczone z danych, ale musz by okrelone przez osob projektujc model.

2. Rola hiperparametrw:
   - Kontroluj zachowanie i struktur modelu uczenia maszynowego.
   - Wpywaj na proces uczenia, szybko konwergencji i zdolno modelu do generalizacji.

3. Przykady hiperparametrw dla rnych algorytmw:

   a) Dla sieci neuronowych:
      - Liczba warstw ukrytych i neuronw w kadej warstwie
      - Wspczynnik uczenia (learning rate)
      - Funkcje aktywacji
      - Wspczynnik regularyzacji
      - Rozmiar partii (batch size)
      - Liczba epok

   b) Dla drzew decyzyjnych i lasw losowych:
      - Maksymalna gboko drzewa
      - Minimalna liczba prbek wymagana do podziau wza
      - Liczba drzew w lesie losowym

4. Metody doboru hiperparametrw:

   a) Rczny dobr:
      - Wymaga duej wiedzy dziedzinowej i dowiadczenia
      - Czsto jest procesem czasochonnym i iteracyjnym

   b) Przeszukiwanie siatki (Grid Search):
      - Systematyczne przeszukiwanie predefiniowanej siatki wartoci hiperparametrw
      - Gwarantuje znalezienie najlepszej kombinacji, ale moe by obliczeniowo kosztowne

   c) Przeszukiwanie losowe (Random Search):
      - Losowe prbkowanie z przestrzeni hiperparametrw
      - Czsto bardziej efektywne ni Grid Search, szczeglnie gdy nie wszystkie hiperparametry s rwnie wane

5. Wpyw hiperparametrw na wydajno modelu:
   - Underfitting vs. Overfitting: Niewaciwy dobr hiperparametrw moe prowadzi do niedopasowania lub przeuczenia modelu
   - Kompromis midzy zoonoci modelu a jego zdolnoci do generalizacji
   - Wpyw na czas treningu i wymagania obliczeniowe

6. Dobre praktyki w doborze hiperparametrw:
   - Uywanie walidacji krzyowej do oceny wydajnoci modelu dla rnych kombinacji hiperparametrw
   - Rozpoczynanie od szerokiego zakresu wartoci i stopniowe zawanie
   - Uwzgldnienie ogranicze obliczeniowych i czasowych przy wyborze metody doboru hiperparametrw

7. Narzdzia i biblioteki do optymalizacji hiperparametrw:
   - Scikit-learn: GridSearchCV, RandomizedSearchCV
   - Optuna: Zaawansowana biblioteka do optymalizacji hiperparametrw
   - Keras Tuner: Dla modeli deep learning w Keras/TensorFlow

#### Przykady hiperparametrw dla rnych algorytmw uczenia maszynowego w jzyku Python

1. Regresja liniowa:

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression(
    fit_intercept=True,  # Czy dopasowa wyraz wolny
    normalize=False,     # Czy normalizowa cechy
    n_jobs=-1            # Liczba rdzeni CPU do uycia (-1 oznacza wszystkie)
)
```

2. Drzewo decyzyjne:

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(
    criterion='gini',     # Funkcja do mierzenia jakoci podziau ('gini' lub 'entropy')
    max_depth=None,       # Maksymalna gboko drzewa
    min_samples_split=2,  # Minimalna liczba prbek wymagana do podziau wza
    min_samples_leaf=1,   # Minimalna liczba prbek wymagana w liciu
    max_features=None,    # Liczba cech do rozwaenia przy szukaniu najlepszego podziau
    random_state=42       # Ziarno losowoci dla powtarzalnoci wynikw
)
```

3. Random Forest:

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(
    n_estimators=100,     # Liczba drzew w lesie
    criterion='gini',     # Funkcja do mierzenia jakoci podziau
    max_depth=None,       # Maksymalna gboko drzewa
    min_samples_split=2,  # Minimalna liczba prbek wymagana do podziau wza
    min_samples_leaf=1,   # Minimalna liczba prbek wymagana w liciu
    max_features='auto',  # Liczba cech do rozwaenia przy szukaniu najlepszego podziau
    bootstrap=True,       # Czy uywa bootstrappingu przy budowie drzew
    n_jobs=-1,            # Liczba rdzeni CPU do uycia
    random_state=42       # Ziarno losowoci
)
```

4. Support Vector Machine (SVM):

```python
from sklearn.svm import SVC

model = SVC(
    C=1.0,                # Parametr regularyzacji
    kernel='rbf',         # Typ jdra ('linear', 'poly', 'rbf', 'sigmoid')
    degree=3,             # Stopie wielomianu (dla kernel='poly')
    gamma='scale',        # Wspczynnik jdra dla 'rbf', 'poly' i 'sigmoid'
    class_weight=None,    # Wagi klas
    probability=False,    # Czy wczy estymacj prawdopodobiestwa
    random_state=42       # Ziarno losowoci
)
```

5. K-Najbliszych Ssiadw (KNN):

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(
    n_neighbors=5,        # Liczba ssiadw do rozwaenia
    weights='uniform',    # Wagi punktw ('uniform' lub 'distance')
    algorithm='auto',     # Algorytm do obliczania najbliszych ssiadw
    leaf_size=30,         # Rozmiar licia (wpywa na szybko budowy drzewa)
    p=2,                  # Parametr potgi dla metryki Minkowskiego
    metric='minkowski',   # Metryka do obliczania odlegoci
    n_jobs=-1             # Liczba rdzeni CPU do uycia
)
```

6. Gradient Boosting:

```python
from sklearn.ensemble import GradientBoostingClassifier

model = GradientBoostingClassifier(
    n_estimators=100,     # Liczba etapw boosting
    learning_rate=0.1,    # Szybko uczenia si
    max_depth=3,          # Maksymalna gboko drzew
    min_samples_split=2,  # Minimalna liczba prbek wymagana do podziau wza
    min_samples_leaf=1,   # Minimalna liczba prbek wymagana w liciu
    subsample=1.0,        # Cz prbek do uycia w kadej iteracji
    max_features=None,    # Liczba cech do rozwaenia przy szukaniu najlepszego podziau
    random_state=42       # Ziarno losowoci
)
```

**Parametry modelu**: wartoci, ktre model dostosowuje w trakcie uczenia.


Parametry modelu:

1. Definicja:
   - Parametry modelu to zmienne wewntrzne, ktre model ustala podczas procesu treningu.
   - S to wartoci, ktre model dostosowuje, aby lepiej dopasowa si do danych treningowych.

2. Rola parametrw:
   - Definiuj zachowanie modelu i jego zdolno do przewidywania lub klasyfikacji.
   - Reprezentuj "wiedz" nabyt przez model z danych treningowych.

3. Przykady parametrw dla rnych modeli:

   a) Regresja liniowa:
      - Wspczynniki (wagi) dla kadej cechy
      - Wyraz wolny (bias)

   b) Sieci neuronowe:
      - Wagi pocze midzy neuronami
      - Biasy neuronw

   c) Drzewa decyzyjne:
      - Warunki podziau w wzach
      - Wartoci przewidywane w liciach


4. Proces uczenia parametrw:

   a) Inicjalizacja:
      - Parametry s zwykle inicjowane losowo lub zgodnie z okrelon strategi.

   b) Optymalizacja:
      - Podczas treningu parametry s iteracyjnie aktualizowane.
      - Celem jest minimalizacja funkcji straty (loss function).

   c) Algorytmy optymalizacji:
      - Gradient Descent i jego warianty (np. Stochastic Gradient Descent, Mini-batch Gradient Descent)
      - Algorytmy adaptacyjne (np. Adam, RMSprop, Adagrad)

5. Rnice midzy parametrami a hiperparametrami:
   - Parametry s uczone z danych, hiperparametry s ustawiane rcznie lub optymalizowane zewntrznie.
   - Liczba parametrw wpywa na zoono modelu, podczas gdy hiperparametry kontroluj proces uczenia.

6. Wymiar parametrw:
   - Liczba parametrw moe si znacznie rni midzy modelami.
   - Proste modele mog mie kilka parametrw, podczas gdy gbokie sieci neuronowe mog mie miliony.

7. Regularyzacja parametrw:
   - Techniki takie jak L1 (Lasso) i L2 (Ridge) regularyzacja pomagaj kontrolowa wartoci parametrw.
   - Celem jest zapobieganie przeuczeniu poprzez ograniczenie zoonoci modelu.

8. Interpretacja parametrw:
   - W niektrych modelach (np. regresja liniowa) parametry maj bezporedni interpretacj.
   - W bardziej zoonych modelach (np. gbokie sieci neuronowe) interpretacja parametrw moe by trudniejsza.


9. Transfer learning i parametry:
    - W transfer learning, parametry wstpnie wytrenowanego modelu s wykorzystywane jako punkt startowy.
    - Cz parametrw moe by "zamroona", a cz dostrajana do nowego zadania.

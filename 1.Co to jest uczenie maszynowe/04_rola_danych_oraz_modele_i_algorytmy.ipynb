###  Rola danych w uczeniu maszynowym

1. Fundamentalne znaczenie danych:
   - Dane s "paliwem" dla modeli uczenia maszynowego.
   - **Jako i ilo danych czsto maj wikszy wpyw na wydajno modelu ni wybr algorytmu**.

2. Rodzaje danych w ML:

   a) Dane strukturalne: 
      - Tabele, bazy danych
      - atwe do przetwarzania, np. dane finansowe, demograficzne
      
   b) Dane niestrukturalne:
      - Tekst, obrazy, dwik, wideo
      - Wymagaj zaawansowanych technik przetwarzania

3. Przygotowanie danych:

   a) Czyszczenie danych:
      - Usuwanie lub imputacja brakujcych wartoci
      - Obsuga wartoci odstajcych
      - Korekcja bdw i niespjnoci
      
   b) Transformacja danych:
      - Normalizacja i standaryzacja
      - Kodowanie zmiennych kategorycznych
      - Redukcja wymiarowoci (np. PCA, t-SNE)
      
   c) Augmentacja danych:
      - Tworzenie syntetycznych przykadw, szczeglnie w zadaniach z obrazami

4. Podzia danych:

   - Zbir treningowy: do uczenia modelu
   - Zbir walidacyjny: do oceny modelu podczas treningu i strojenia hiperparametrw
   - Zbir testowy: do kocowej oceny modelu

5. Znaczenie reprezentatywnoci danych:
   - Dane treningowe powinny dobrze reprezentowa rzeczywiste przypadki
   - Niezrwnowaone klasy mog prowadzi do stronniczoci modelu

6. Wpyw wielkoci zbioru danych:
   - Wiksze zbiory danych czsto prowadz do lepszych modeli
   - Mae zbiory danych mog prowadzi do przeuczenia
   - Techniki dla maych zbiorw: transfer learning, data augmentation

7. Jako vs. ilo:
   - Czasami mniejszy, ale wysokiej jakoci zbir danych jest lepszy ni duy, ale zaszumiony
   - Znaczenie etykietowania danych i kontroli jakoci

8. Dynamika danych:
   - Dane mog zmienia si w czasie (concept drift)
   - Potrzeba cigego monitorowania i aktualizacji modeli

9. Prywatno i etyka danych:
   - Ochrona danych osobowych (np. RODO)
   - Unikanie stronniczoci i dyskryminacji w danych

10. Eksploracyjna analiza danych (EDA):
    - Kluczowy etap przed modelowaniem
    - Zrozumienie rozkadw, korelacji i wzorcw w danych

11. Feature engineering (inynieria cech):
    - Tworzenie nowych cech na podstawie istniejcych danych
    - Czsto kluczowe dla poprawy wydajnoci modelu

12. Obsuga danych strumieniowych:
    - Przetwarzanie danych w czasie rzeczywistym
    - Techniki uczenia przyrostowego (batch)

13. Dane w rnych paradygmatach uczenia:
    - Uczenie nadzorowane: etykietowane dane
    - Uczenie nienadzorowane: dane bez etykiet
    - Uczenie pnadzorowane: mieszanka danych etykietowanych i nieetykietowanych
    - Uczenie ze wzmocnieniem: dane generowane przez interakcj z rodowiskiem

14. Wyzwania zwizane z danymi:
    - Big Data: obsuga ogromnych iloci danych
    - Rzadkie dane: problemy z niedostateczn reprezentacj niektrych klas
    - Zaszumione dane: radzenie sobie z bdami i nieistotnymi informacjami

15. Narzdzia do pracy z danymi w Pythonie:
    - Pandas: manipulacja i analiza danych
    - NumPy: operacje numeryczne
    - Scikit-learn: preprocessing danych
    - Matplotlib i Seaborn: wizualizacja danych

16. Przechowywanie i dostp do danych:
    - Bazy danych (SQL, NoSQL)
    - Systemy plikw rozproszone (np. Hadoop)
    - Chmurowe rozwizania do przechowywania danych

17. Transfer learning i pre-trained models:
    - Wykorzystanie modeli wytrenowanych na duych zbiorach danych
    - Adaptacja do specyficznych zada z mniejsz iloci danych

### Koncepcja modeli i algorytmw

Model w uczeniu maszynowym to matematyczna reprezentacja procesu decyzyjnego:
- Model jest tworzony przez algorytm uczenia si na podstawie danych treningowych.
- Moe to by np. funkcja liniowa w regresji, drzewo decyzyjne, czy zoona sie neuronowa.

Algorytmy uczenia maszynowego to procedury uywane do tworzenia i trenowania modeli:
- Istniej rne rodzaje algorytmw, dostosowane do rnych typw problemw i danych.
- Przykady obejmuj algorytmy regresji, klasyfikacji, grupowania, czy redukcji wymiarowoci.
- Wybr odpowiedniego algorytmu zaley od natury problemu, dostpnych danych i podanych wynikw.

Proces uczenia:
1. Inicjalizacja modelu z pocztkowymi parametrami.
2. Wykorzystanie danych treningowych do dostosowania parametrw modelu.
3. Iteracyjne poprawianie modelu poprzez minimalizacj funkcji straty (loss function).
4. Walidacja modelu na nowych danych w celu oceny jego zdolnoci do generalizacji.

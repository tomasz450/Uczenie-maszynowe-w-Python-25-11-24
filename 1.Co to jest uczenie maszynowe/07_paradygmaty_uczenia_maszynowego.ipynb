## Paradygmaty uczenia maszynowego

### Uczenie nadzorowane
- Definicja: Proces trenowania modelu na podstawie oznakowanych danych, gdzie dla każdego przykładu wejściowego znana jest pożądana odpowiedź (etykieta).
- Cel: Nauczyć model przewidywania poprawnych etykiet dla nowych, niewidzianych wcześniej danych.
- Kluczowe elementy:
  - Zbiór treningowy: pary (dane wejściowe, etykieta)
  - Funkcja straty: mierzy różnicę między przewidywaniami modelu a rzeczywistymi etykietami
  - Algorytm optymalizacji: dostosowuje parametry modelu, aby zminimalizować funkcję straty
- Przykłady zastosowań:
  - Klasyfikacja: przypisywanie kategorii (np. rozpoznawanie spamu w e-mailach)
  - Regresja: przewidywanie wartości liczbowych (np. prognozowanie cen nieruchomości)
- Popularne algorytmy:
  - Drzewa decyzyjne i lasy losowe
  - Maszyny wektorów nośnych (SVM)
  - Sieci neuronowe
- Wyzwania:
  - Potrzeba dużej ilości oznakowanych danych
  - Ryzyko przeuczenia (overfitting)

### 4.2. Uczenie nienadzorowane
- Definicja: Proces analizy i grupowania nieoznakowanych danych w celu odkrycia ukrytych wzorców lub struktur.
- Cel: Znalezienie nieoczywistych zależności w danych bez wcześniej zdefiniowanych etykiet.
- Kluczowe elementy:
  - Zbiór danych: nieoznakowane przykłady
  - Miary podobieństwa lub odległości między punktami danych
  - Kryteria oceny jakości grupowania lub redukcji wymiarowości
- Przykłady zastosowań:
  - Grupowanie klientów w segmenty marketingowe
  - Wykrywanie anomalii w transakcjach finansowych
  - Kompresja danych i redukcja wymiarowości
- Popularne algorytmy:
  - K-means i hierarchiczne metody grupowania
  - DBSCAN (grupowanie oparte na gęstości)
  - Analiza głównych składowych (PCA)
  - Autoenkodery
- Wyzwania:
  - Trudność w ocenie jakości wyników
  - Interpretacja odkrytych wzorców

### 4.3. Uczenie ze wzmocnieniem
- Definicja: Proces, w którym agent uczy się optymalnych działań poprzez interakcję ze środowiskiem i otrzymywanie nagród lub kar.
- Cel: Znalezienie strategii (polityki) maksymalizującej skumulowaną nagrodę w długim okresie.
- Kluczowe elementy:
  - Agent: podmiot podejmujący decyzje
  - Środowisko: przestrzeń, w której agent działa
  - Stany: sytuacje, w których może znaleźć się agent
  - Akcje: możliwe działania agenta
  - Nagrody: informacja zwrotna o jakości podjętych decyzji
  - Polityka: strategia wyboru akcji w danym stanie
- Przykłady zastosowań:
  - Gry (np. AlphaGo)
  - Robotyka i autonomiczne pojazdy
  - Optymalizacja systemów rekomendacji
- Popularne algorytmy:
  - Q-learning
  - Policy Gradient
  - Deep Q-Network (DQN)
  - Proximal Policy Optimization (PPO)
- Wyzwania:
  - Kompromis między eksploracją a eksploatacją
  - Długi czas uczenia w złożonych środowiskach

### 4.4. Uczenie półnadzorowane
- Definicja: Hybrydowe podejście łączące cechy uczenia nadzorowanego i nienadzorowanego, wykorzystujące zarówno oznakowane, jak i nieoznakowane dane.
- Cel: Poprawa wydajności modeli uczenia nadzorowanego poprzez wykorzystanie dodatkowych, nieoznakowanych danych.
- Kluczowe elementy:
  - Mały zbiór oznakowanych danych
  - Duży zbiór nieoznakowanych danych
  - Metody propagacji etykiet lub regularyzacji
- Przykłady zastosowań:
  - Klasyfikacja tekstu z ograniczoną liczbą oznakowanych dokumentów
  - Rozpoznawanie obrazów z częściowo oznakowanym zbiorem
- Wyzwania:
  - Zapewnienie spójności między oznakowanymi i nieoznakowanymi danymi
  - Unikanie propagacji błędów w procesie pseudo-etykietowania

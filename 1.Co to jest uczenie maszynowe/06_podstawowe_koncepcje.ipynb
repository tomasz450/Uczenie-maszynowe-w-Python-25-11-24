## Podstawowe koncepcje

### Cechy (features) i etykiety (labels)

Cechy (features):
- Definicja: Cechy to mierzalne właściwości lub charakterystyki obserwacji, które są używane jako dane wejściowe do modelu uczenia maszynowego.
- Znaczenie: Cechy są kluczowe dla skuteczności modelu, ponieważ dostarczają informacji, na podstawie których model podejmuje decyzje lub dokonuje predykcji.
- Typy cech:
  * Numeryczne: ciągłe (np. wzrost) lub dyskretne (np. liczba pokoi)
  * Kategoryczne: nominalne (np. kolor) lub porządkowe (np. ocena w skali)
  * Binarne: przyjmujące dwie wartości (np. tak/nie)
  * Tekstowe: ciągi znaków wymagające przetworzenia
- Inżynieria cech: Proces tworzenia, wyboru i transformacji cech w celu poprawy wydajności modelu.

Etykiety (labels):
- Definicja: Etykiety to zmienne docelowe, które model stara się przewidzieć lub sklasyfikować.
- Zastosowanie: Używane głównie w uczeniu nadzorowanym, gdzie model uczy się mapowania między cechami a etykietami.
- Typy problemów:
  * Klasyfikacja: etykiety są kategoriami (np. spam/nie spam)
  * Regresja: etykiety są wartościami ciągłymi (np. cena domu)
- Znaczenie: Jakość i dokładność etykiet ma kluczowe znaczenie dla skuteczności modelu.

### Zbiory treningowe, walidacyjne i testowe

Zbiór treningowy:
- Definicja: Część danych używana do trenowania modelu.
- Cel: Model uczy się na tych danych, dostosowując swoje parametry.
- Wielkość: Zwykle największa część dostępnych danych (np. 60-80%).

Zbiór walidacyjny:
- Definicja: Część danych używana do oceny modelu podczas treningu i strojenia hiperparametrów.
- Cel: Pomaga w wyborze najlepszego modelu i zapobiega przeuczeniu.
- Wielkość: Mniejsza niż zbiór treningowy (np. 10-20% danych).

Zbiór testowy:
- Definicja: Część danych używana do końcowej oceny wydajności modelu.
- Cel: Symuluje rzeczywiste, niewidziane wcześniej dane.
- Wielkość: Zwykle najmniejsza część (np. 10-20% danych).
- Znaczenie: Dostarcza niezależnej oceny generalizacji modelu.

Techniki podziału danych:
- Losowy podział
- Stratyfikowany podział (zachowuje proporcje klas)
- Walidacja krzyżowa (cross-validation)

### Underfitting i overfitting

Underfitting (niedopasowanie):
- Definicja: Model jest zbyt prosty, aby uchwycić złożoność danych.
- Symptomy:
  * Słaba wydajność zarówno na zbiorze treningowym, jak i testowym
  * Wysoki błąd (bias) modelu
- Przyczyny:
  * Zbyt prosty model
  * Niewystarczająca liczba cech
  * Zbyt krótki czas treningu
- Rozwiązania:
  * Zwiększenie złożoności modelu
  * Dodanie nowych, istotnych cech
  * Wydłużenie czasu treningu

Overfitting (przeuczenie):
- Definicja: Model jest zbyt złożony i "zapamiętuje" szum w danych treningowych zamiast uczyć się ogólnych wzorców.
- Symptomy:
  * Doskonała wydajność na zbiorze treningowym, ale słaba na zbiorze testowym
  * Wysoka wariancja modelu
- Przyczyny:
  * Zbyt złożony model w stosunku do ilości danych
  * Zbyt długi czas treningu
  * Szum w danych treningowych
- Rozwiązania:
  * Regularyzacja (L1, L2)
  * Dropout (w sieciach neuronowych)
  * Zwiększenie ilości danych treningowych
  * Uproszczenie modelu
  * Wczesne zatrzymywanie (early stopping)

Kompromis bias-variance:
- Bias: Błąd wynikający z założeń modelu
- Variance: Wrażliwość modelu na fluktuacje w danych treningowych
- Cel: Znalezienie optymalnego punktu między underfittingiem a overfittingiem

Monitorowanie i zapobieganie:
- Krzywe uczenia (learning curves): Wizualizacja wydajności modelu na zbiorach treningowym i walidacyjnym w trakcie treningu
- Walidacja krzyżowa: Technika oceny modelu na różnych podzbiorach danych
- Regularyzacja: Techniki ograniczające złożoność modelu
